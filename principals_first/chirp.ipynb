```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametreler
f_start = 24e9  # Başlangıç frekansı (24 GHz)
f_stop = 24.2e9  # Bitiş frekansı (24.2 GHz)
B = f_stop - f_start  # Bant genişliği: 200 MHz
T_chirp = 1e-3  # Chirp süresi: 1 ms
c = 3e8  # Işık hızı

# Hedef mesafesi
R_target = 100  # 100 metre

# Zaman vektörü
t = np.linspace(0, T_chirp, 1000)

# Frekans modülasyonu (lineer chirp)
f_t = f_start + (B / T_chirp) * t

# Gidiş-dönüş gecikmesi
delay = 2 * R_target / c

# Gönderilen ve alınan sinyallerin frekans farkı
f_beat = (B / T_chirp) * delay
```

```python
print(f"Bant genişliği: {B/1e6:.1f} MHz")
print(f"Mesafe çözünürlüğü: {c/(2*B):.2f} metre")
print(f"Beat frekansı: {f_beat/1e3:.2f} kHz")
print(f"Hesaplanan mesafe: {(f_beat * c * T_chirp)/(2*B):.1f} metre")
```

```python
# Grafik
plt.figure(figsize=(10, 4))
plt.plot(t*1e3, f_t/1e9, 'b-', label='Gönderilen sinyal')
plt.plot((t+delay)*1e3, f_t/1e9, 'r--', label='Alınan sinyal (gecikmeli)')
plt.xlabel('Zaman (ms)')
plt.ylabel('Frekans (GHz)')
plt.legend()
plt.grid(True)
plt.title('FMCW Radar Chirp Sinyali')
plt.show()
```
